# chi-square test

- 카이제곱의 검정통계값이 커질수록 영가설 기각에 더 강한 증거를 보여준다.

    모든 관측값과 기대값이 같으면 카이제곱은 0일 것이다.
    이 둘의 차이(f와 e)가 커질수록 카이제곱이 커지고 이는 영가설 기각의 강력한 증거이다          

- 교차표가 n개의 행과 m개의 열로 구성되어 있을 때,
기대빈도가 모든 범주에서 5 이상이라면
검정통계량은 자유도가 (n − 1)(m − 1)인 카이제곱 분포를 따른다  

- 카이제곱 분포는 one tail 분포이다
오직 양수만 갖는다                                    
카이제곱 값이 커지면 p값은 더 작아진다.         

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Load the dataset
soju = pd.read_csv('/content/drive/MyDrive/data analysis and coding/sojulove.csv')
```

```python
# Create a contingency table
contingency_table=pd.crosstab(soju['gender'].sort_values(ascending=True), soju['spreference'].sort_values(ascending=True))
contingency_table.columns=["Grapefruit", "Isback", "Fresh"] # column label
contingency_table.index=["Male", "Female"] # row lable
print("Observed Frequencies:")
print(contingency_table,'\n')
```

```python
# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
# Print the results
print("Chi-Square Test Results")
print(chi2)
print(p)
print(dof ,'\n')
```

```python
# Convert the expected frequencies to a DataFrame for better readability
expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)
print("Expected Frequencies:")
print(expected_df,'\n')
```
