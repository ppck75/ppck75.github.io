# **파이썬 코드 작성 원리**

* 코드의 기본 구성 요소: **객체(object), 함수(function), 인자(argument)**
* 코딩 원리: **객체.함수(인자)**  
  → 특정 자료(객체)에 대해 함수를 적용하고, 필요 시 인자(추가 정보)를 전달하여 동작 수행

* **객체(object)**: 컴퓨터 메모리상에 존재하는 모든 것 (예: 자료형, 함수, 라이브러리 등)  
  - 객체를 생성하면 컴퓨터 메모리의 특정 공간이 할당됨

* **변수(variable)**: 객체를 가리키는 이름(참조)  
  - 흔히 "객체를 담고 있다"라고 비유하기도 함

* **함수(function)**: 객체에 대해 특정 동작을 수행하는 도구  
  - 예: `print(a)` → 변수 `a` 안에 있는 객체를 출력하는 함수

* **인자(argument)**: 함수 실행 시 필요한 정보나 조건  
  - 예:  
    ```python
    text = "Hello Python World"
    text.split(" ")
    ```

  - 예: `print()`, `len()`, `type()`

* **주석(comments)**: 코드 내 설명/메모를 작성할 때 사용  
  - 실행 시 무시됨

# **객체 생성과 변수 선언, 그리고 다양한 자료형**

* **객체(object)**: 파이썬에서 다루는 모든 것 (예: `1`, `"1"`)  
  - 파이썬에서 값이 생성되면 메모리에 객체로 저장됨

* **변수(variable)**: 객체(값)를 가리키는 이름  
  - 예: `x = 1` → 변수 `x`는 정수 객체 `1`을 참조함

* 객체의 자료는 다양한 **자료형(data type)** 으로 존재 (총 7가지 기본형)  
  - **기본 자료형**: 문자열(`str`), 숫자형(`int`, `float`), 불형(`bool`)  
  - **복합 자료형**: 리스트(`list`), 사전(`dict`), 튜플(`tuple`), 집합(`set`)  

* 같은 값이라도 **자료형에 따라 결과가 달라짐**  
  - 예: `"1" + "1"` → `"11"` (문자열 연결)  
  - 예: `1 + 1` → `2` (숫자 연산)

* 객체의 자료형은 **type() 함수**로 확인 가능  
  - 예:  
    ```python
    type(1)        # <class 'int'>
    type("1")      # <class 'str'>
    type(True)     # <class 'bool'>
    ```


### 객체 생성(Instantiation)

````python
# <객체 생성> (Instantiation)
# "세상에 없던 값을 메모리에 올려놓는 것" = 객체 생성
# 객체는 메모리 상의 실제 데이터

1         # 숫자 1 자체도 하나의 "객체" (정수형 int)
# a       # 변수 선언 전에는 사용할 수 없음 → NameError 발생

a = 1     # 정수 1 객체를 생성하고, 변수 a가 그것을 참조
# '=' 는 "할당 연산자"임 (비교 연산자 '==' 와 다름)

print(a)               # 변수 a가 가리키는 객체 출력
print(type(a))         # a의 자료형 확인

# id() 함수: 객체가 저장된 메모리 주소(고유 ID) 확인
print(id(a))

# hex() 함수로 메모리 주소를 16진수로 변환하여 출력
print(hex(id(a)))
````

### 변수 선언(Declaration)

````python
# <변수 선언> (Declaration)
# 변수는 "객체를 가리키는(참조하는) 이름"
# 파이썬에서는 변수만 따로 선언하는 방식은 없음
# 반드시 객체(값)를 먼저 만들고, 그 객체를 변수에 할당해야 함

b        # 에러 발생 → NameError: name 'b' is not defined
# 변수 자체는 "이름"이지, 객체가 아니기 때문

"b"      # 문자열 객체
b = "b"  # 문자열 객체 "b"를 생성하고, 변수 b가 그것을 참조
print(b)

# type() 으로 자료형 확인
print(type(b))   # <class 'str'>

# id() : 객체의 고유 ID 값 확인
print(id(b))

# hex(id()) : 객체의 메모리 주소를 16진수로 변환
print(hex(id(b)))
````

### **문자열 인덱싱(Indexing)과 슬라이싱(Slicing)**

* 대괄호 `[ ]` 를 사용해 문자열의 특정 요소(문자)를 가리키거나 잘라낼 수 있음  
* **파이썬은 인덱스를 0부터 센다!**

* **인덱싱(Indexing)**  
  - 문자열의 특정 위치에 있는 문자를 가져오기  
  - 예: `text[3]`

* **슬라이싱(Slicing)**  
  - 문자열의 일부 범위를 잘라내기  
  - 예: `text[n1:n2]` → `n1` 이상 ~ `n2` 미만 구간을 추출

* **음수 인덱스 (negative index)**  
  - 뒤에서부터 번호를 매김  
  - `-1` = 마지막 문자, `-2` = 뒤에서 두 번째 문자  

* **슬라이싱 변형 패턴**  
  - `[:n]` → 처음부터 `n-1` 인덱스까지  
  - `[n:]` → `n`번째부터 끝까지  
  - `[::step]` → step 간격으로 추출 (건너뛰기)  

````python
# 1. 문자열 인덱싱: 문자열 요소 가리키기
b = "KHU"
print(b[-1])   # U (뒤에서 첫 번째 문자)

# 여러 줄 문자열 (docstring 예제)
s_long = '''I don't know what we've done
되돌아가긴 싫어, もう知っている
Don't know what we've been sold
見つけられるよ, so it's sure (come on)
'''

# 문자열 인덱싱
print(s_long[0])    # I (첫 번째 문자)
print(s_long[1])    #   (공백)
print(s_long[-1])   # \n (마지막 줄바꿈 문자)

# 음수 인덱스 예시
print(s_long[-2])   # ) (뒤에서 두 번째 문자)
print(s_long[-5])   # 공백 (뒤에서 다섯 번째 문자)

# 문자열 길이 확인
print(len(s_long))  # 111 (공백·줄바꿈 포함한 전체 문자 수)
````

````python
# 2. 문자열 슬라이싱: 문자열 요소 잘라내기

# 개별 문자 이어붙이기 vs 슬라이싱
print(s_long[0] + s_long[1] + s_long[2] + s_long[3])   # I do
print(s_long[0:4])   # I do (0 이상 4 미만)

# 기본 슬라이싱
print(s_long[:10])   # 1번째 숫자부터 10번째 숫자까지 나온다.  (I don't kn)
s_long[10]
# 슬라이싱 변형 패턴
print(s_long[:5])    # 처음부터 4번째까지 → I don
print(s_long[5:])    # 5번째부터 끝까지
print(s_long[::2])   # 2칸씩 건너뛰며 추출
print(s_long[::-1])  # 문자열 뒤집기

# 슬라이싱 이후에도 원본은 변하지 않음 (불변성)
print(s_long)        # 원본 문자열 그대로 유지   #지피티: 리스트는 가변 자료형이지만, 슬라이싱을 하는 행위 자체는 원본을 변경하지 않는다. 하지만, 슬라이싱 영역에 세개의 값을 대입하는게 리스트에서는 가능하다!! 문자열은 불가능
````

### **문자열 관련 주요 함수**

* 파이썬 문자열은 **내장 함수**를 통해 다양한 처리가 가능 (별도 라이브러리 필요 없음)  
* 문자열은 **immutable(불변) 자료형** → 원본 문자열은 바뀌지 않고 **새로운 문자열**을 반환함  

* 주요 함수 예시  
  - `len()` : 문자열 길이 반환  
  - `type()` : 자료형 확인  
  - `count()` : 특정 문자열이 등장하는 횟수  
  - `find()` : 특정 문자열이 처음 나타나는 위치 (없으면 -1 반환)  
  - `index()` : 특정 문자열이 처음 나타나는 위치 (없으면 에러 발생)  
  - `upper()` : 모두 대문자로 변환  
  - `lower()` : 모두 소문자로 변환  
  - `strip()` : 앞뒤 공백 제거  
  - `split()` : 구분자를 기준으로 문자열 나누기 → 리스트 반환  
  - `join()` : 리스트 요소를 특정 구분자로 이어붙이기  
  - `replace()` : 특정 문자열을 다른 문자열로 교체


````python
s_long = '''I don't know what we've done
되돌아가긴 싫어, もう知っている
Don't know what we've been sold
見つけられるよ, so it's sure (come on)
'''
print(s_long)
s_long.find('w') # w는 12번째에 있으니까 11이 나온다.
# (1) count: 부분 문자열의 "겹치지 않는" 등장 횟수 (대/소문자 구분)
print(s_long.count("we"))     # 예: 'we'가 몇 번 나오는지

# (2) find와 index: 부분 문자열의 "첫 등장 위치" 반환
print(s_long.find("I"))       # 없으면 -1 (안전하게 검사할 때 유용)
# print(s_long.index("not-exist"))  # 없으면 ValueError 발생 → 주석 처리
print(s_long.index("we"))     # 존재할 때는 find와 동일한 결과

# (3) lower / upper: 대소문자 변환 (라틴 문자 중심, 한글/일본어는 변화 없음)
print("Hello".lower())        # hello
print("Hello".upper())        # HELLO
print(s_long.upper())         # 원본은 변하지 않음 (문자열은 불변)

# (4) join: '구분자'.join(문자열들의_시퀀스)
#    - 문자들 사이에 기호를 끼워 넣어 합침
#    - join의 인자는 "문자열들의 시퀀스(list/tuple 등)" 여야 함
print("/".join(s_long))       # 각 문자 사이에 '/' (문자 단위 join)
words = ["KHU", "Python", "NewJeans"]
print(" ♡ ".join(words))      # KHU ♡ Python ♡ NewJeans (단어 단위 join)

# (5) strip: 양쪽 공백/개행 제거 (기본 인자 없음)
print(s_long.strip())
print("  안녕하세요  ".strip())
# 변경을 유지하려면 재할당 필요
s = s_long.strip()

# (6) replace: 특정 문자열을 다른 문자열로 치환 (새 문자열 반환)
print(s_long.replace("we", "I"))
print(s_long.replace("되돌아가긴", "이대로있긴"))
print(s_long.replace("\n", " "))  # 줄바꿈을 공백으로 (전처리에 매우 유용)

# 원본 확인 (불변성: 원본은 그대로)
print(s_long)

# 불변성 때문에 변경 유지하려면 재할당
s_long_1 = s_long.replace("\n", " ")
print(s_long_1)

# (7) split: 문자열을 "구분자" 기준으로 나눠 리스트 반환
print(s_long.split(" "))              # 공백 1개 기준 → 연속 공백은 빈 문자열 포함
print(s_long.replace("\n", " ").split(" "))  # 줄바꿈 제거 후 공백 기준 분리

# 권장: 인자를 생략하면 "연속 공백/개행/탭"을 모두 공백으로 간주하여 깔끔히 분리
print(s_long.split())                 # 데이터 전처리 시 가장 실용적

# 전처리 체이닝 예시:
# 1) 줄바꿈을 공백으로 → 2) 앞뒤 공백 제거 → 3) 공백 기준 분리
tokens = s_long.replace("\n", " ").strip().split()
print(tokens)

# (8) 기타 내장 함수
print(type(s_long))   # <class 'str'>
print(len(s_long))    # 전체 문자 수(공백/개행 포함)
print(round(3.5678, 2))  # 3.57   3.5678을 두 자리까지 반올림해줘
````

````python
# 문자열 포맷팅 예제

s_1 = '뉴진스가 "안녕하세요"라고 인사했다'
print(s_1)

# (1) format() 함수
s_2 = '뉴진스가 "{}"라고 인사했다'.format("사랑해요")
print(s_2)

s_3 = '{}가 "{}"라고 인사했다'.format("하니", "사랑해요")
print(s_3)   # ← 수정: s_3 출력

# {}에 할당할 내용을 변수로 전달
a = "하니"
b = "사랑해요"
s_4 = '{}가 "{}"라고 인사했다'.format(a, b)
print(s_4)

# 숫자형 포매팅
s_5 = "{} 더하기 {} 은 {}".format(3, 4, 7)
print(s_5)
````

 ## 숫자형(numeric)


*   숫자 형태로 이루어진 자료형

*   예: 1, 2, 3, 3,14, ...

*   정수형(Integer)와 실수형(Floating-point)으로 구성

*   int, float 로 자료형 클래스를 각각 표현(type()으로 확인)

*   숫자연산자: +, -, *, /, **, //, % 등 사용가능




````python
## 숫자형 (numeric)

# 정수형과 실수형
i = 5
print(i, type(i))   # 5 <class 'int'>

i = "5"
print(i, type(i))   # "5" <class 'str'>

# 문자열 "5" → 정수 5로 변환
i = int("5")
print(i, type(i))   # 5 <class 'int'>

# 문자열 "2.5" → 실수 2.5로 변환
f = float("2.5")
print(f, type(f))   # 2.5 <class 'float'>
````

 ### 숫자 연산

````python
print(5 + 5)   # 10 (더하기)
print(5 - 5)   # 0 (빼기)
print(10 / 2)  # 5.0 (나누기 → 항상 float)
print(5 * 2)   # 10 (곱하기)
print(3 ** 2)  # 9 (거듭제곱)
print(7 // 3)  # 2 (몫)
print(7 % 3)   # 1 (나머지)
````

### 축약 연산자 (할당 연산자): 본래 값에 일정량 더하기

````python
x = 2
x += 1      # x = x + 1
print(x)    # 3

x = 2
x -= 1      # x = x - 1
print(x)    # 1

x = 2
x *= 3      # x = x * 3
print(x)    # 6

x = 10
x /= 4      # x = x / 4
print(x)    # 2.5
````

## 불(bool): 논리자료형

* 참(True)과 거짓(False)을 나타내는 자료형
* True와 False 2가지 값만 가짐 (파이썬 키워드) - 꼭 첫글자(T or F)는 대문자
* 모든 조건문의 기반

````python
## 불(bool): 논리자료형

# True / False (조건식의 결과)
print(2 > 1)            # True (2가 1보다 큼)
print(type(2 > 1))      # <class 'bool'>

print("2 > 1")

# 논리 연산자: and, or, not
print((2 > 1) and (0 > 1))  # False (앞은 True, 뒤는 False → 둘 다 참이 아니므로 False)
print((2 > 1) or (0 > 1))   # True  (앞은 True, 뒤는 False → 하나라도 참이면 True)
print(not (2 > 1))          # False (not True → False)
````

````python
# in 연산자: 포함 여부 확인
a = "안녕하세요. 뉴진스입니다"
print("뉴진스" in a)    # True ("뉴진스"가 문자열 a 안에 있니?)
print("하니" in a)      # False ("하니"는 있니?)
print("입니" in a)      # True ("입니"는 있니?)

# if 문에서 활용 예시
if "뉴진스" in a:   # 조건식이 True일 때 실행
    print("a 문자열 안에 뉴진스가 있습니다")
else:
    print("a 문자열 안에 뉴진스가 없습니다")
````

## 리스트(list)

* 여러 개의 요소(문자열, 숫자, 불 등)를 담을 수 있는 자료형  
* 생성 방법: 대괄호 `[ ]` 로 감싸고, 요소들은 쉼표(,)로 구분  
* 리스트는 **순서(index)** 를 가지며, 수정/삭제가 가능 (mutable 자료형)



````python
# 리스트 생성 예시

# (1) 빈 리스트
lst_null = []                # 빈 리스트 생성 (중요)
lst_null2 = list()           # 같은 결과 (list() 함수 사용)

# (2) 숫자로 이루어진 리스트
lst_numeric = [0, 1, 2, 3, 4]
lst_height = [162, 165, 169, 170, 165]   # 키 데이터
lst_year   = [2004, 2006, 2004, 2008, 2005]  # 출생 연도 데이터

# (3) 문자열로 이루어진 리스트
lst_string  = ["하니", "해린", "민지", "혜인", "다니엘"]
lst_string1 = ["Hanni", "Haerin", "Minji", "Hyein", "Danielle"]
lst_string2 = ["베트남&호주", "한국", "한국", "한국", "호주"]  # 국적 데이터

# (4) 불(bool) 자료형 리스트
lst_bool = [True, False, True, False]

# (5) 혼합 자료형 리스트
lst_mix = ["하니", "해린", 2004, 2006]   # 문자열 + 숫자 혼합

# (6) 리스트 안에 리스트 (중첩 리스트)
lst_lst = ["하니", "해린", [162, 165]]   # 리스트 요소로 또 다른 리스트 포함


# 출력 예시
print(lst_null)
print(lst_numeric)
print(lst_string)
print(lst_bool)
print(lst_mix)
print(lst_lst)
````

### 리스트 인덱싱과 슬라이싱

*   리스트 인덱싱(indexing): 리스트의 각 요소는 인덱스를 통해 접근 가능(형식: 리스트이름[위치])

*   요소의 위치는 0부터 시작

*   리스트 슬라이싱(slicing): 리스트의 부분집합을 추출(형식: 리스트이름[start:end])


````python
# 리스트 준비
lst_string = ["하니", "해린", "민지", "혜인", "다니엘"]


# 1. 리스트 인덱싱 (리스트 요소 가리키기)

# 2번째 요소 호출 (0부터 시작 → "해린")
print(lst_string[1])    # 해린

# 마지막 요소 호출
print(lst_string[-1])   # 다니엘

# 마지막 요소를 "대니"로 수정 (리스트는 가변형 → mutable)
lst_string[-1] = "대니"
print(lst_string)       # ['하니', '해린', '민지', '혜인', '대니']
````

````python
# 2. 리스트 슬라이싱 (리스트 요소 잘라내기)

# 2번째 ~ 4번째 요소 (인덱스 1~3)
print(lst_string[1:4])  # ['해린', '민지', '혜인']

# 4번째 요소부터 끝까지
print(lst_string[3:])   # ['혜인', '대니']

# 리스트 길이 구하기 (len 함수)
print(len(lst_string))  # 5
del lst_string[-1]
````

````python
lst_string[1:4]   # lst_string 리스트에서 두번째에서 네번째 요소까지 3개 요소 호출.
lst_string[3:]    # lst_string 리스트에서 네번째 요소부터 마지막 요소까지 호출.
````

### 리스트 요소 수정하기와 삭제하기 (가능!)

* 리스트는 **가변 자료형(mutable)**  
  - 사전(dict), 집합(set)도 가변 자료형  
  - 문자열(str), 튜플(tuple), 숫자(numeric), 논리자료형(bool) 등은 불변 자료형(immutable)

* 장점: 데이터 크기가 크고, 구조 특성상 **자주 수정·삭제가 필요할 때 유리**

* 인덱싱(indexing): 리스트의 각 요소는 인덱스를 통해 접근 가능 (0부터 시작)

* 슬라이싱(slicing): 리스트의 부분집합을 추출 (형식: `리스트이름[start:end]`)


````python
lst_string = ["하니", "해린", "민지", "혜인", "다니엘"]

# (1) 요소 수정 (재할당 없이 기존 리스트 변경됨 → 리스트는 mutable)
lst_string[-1] = "카리나"
print(lst_string)   # ['하니', '해린', '민지', '혜인', '카리나']

# (2) 요소 삭제 - remove() 함수 (리스트에서 '민지' 삭제)
lst_string.remove("민지")
print(lst_string)   # ['하니', '해린', '혜인', '카리나']

# (3) 요소 삭제 - del 키워드 (2번째 요소 삭제: '해린')
del lst_string[1]
print(lst_string)   # ['하니', '혜인', '카리나']
````

````python
# 리스트 준비
lst_string = ["하니", "해린", "민지", "혜인", "다니엘"]

# (1) 리스트 길이 구하기
print(len(lst_string))
# 5 → 리스트에 5개의 요소가 있음

# (2) 문자열 길이 구하기
name = "뉴진스"
print(len(name))
# 3 → '뉴', '진', '스' 3글자
````

### 리스트 관련 주요 메서드(내장 함수)

* 리스트 자료를 유연하게 조작할 수 있음  
* 자주 쓰이는 메서드
  - `append()` : 리스트 맨 뒤에 요소 추가
  - `extend()` : 리스트 뒤에 다른 리스트(또는 시퀀스)의 모든 요소 추가
  - `insert()` : 원하는 위치에 요소 삽입
  - `remove()` : 특정 값 삭제 (첫 번째 등장만)
  - `pop()` : 특정 위치 요소를 꺼내면서 삭제 (기본은 마지막)
  - `sort()` : 리스트 요소 정렬
  - `reverse()` : 리스트 요소 역순으로 뒤집기

````python
lst_string = ["하니", "해린", "민지", "혜인", "다니엘"]

# 1. append(): 리스트 맨 뒤에 요소 추가
lst_string.append("방시혁")
print(lst_string)

# 2. sort(): 리스트 정렬하기
lst_string.sort()                 # 오름차순 정렬
print(lst_string)

lst_string.sort(reverse=True)     # 내림차순 정렬
print(lst_string)

print(sorted(lst_string))         # sorted() 함수도 가능 (원본은 유지)

# 3. index(): 특정 요소의 위치(인덱스) 반환
print(lst_string.index("하니"))   # 요소 "하니"의 위치
# print(lst_string.index("카리나"))  # 없으면 오류(ValueError)

# 4. count(): 특정 요소의 개수 세기
print(lst_string.count("다니엘"))

# 5. insert(): 특정 위치에 요소 삽입
lst_string.insert(0, "버니즈")    # 0번 위치에 "버니즈" 삽입, 나머지 요소는 뒤로 밀림
print(lst_string)

# 6. remove(): 특정 요소 제거 (첫 번째 등장만 삭제)
lst_string.remove("버니즈")
lst_string.remove("방시혁") # 요소가 없는데?
print(lst_string)

# 7. extend(): 리스트 확장 (다른 리스트 요소들을 뒤에 추가)
lst_string.extend(["버니즈1", "버니즈2", "버니즈3"])
print(lst_string)

# 리스트 이어붙이기(+) 연산도 가능
print(lst_string + ["추가1", "추가2"])

# 8. reverse(): 리스트 뒤집기
lst_string.reverse()
print(lst_string)

# 9. pop(): 요소 꺼내면서 삭제
print(lst_string.pop())    # 마지막 요소 꺼내기
print(lst_string.pop(0))   # 첫 번째 요소 꺼내기
print(lst_string)

# 10. join(): 리스트 요소들을 문자열로 합치기 (문자열 전용 메서드)
print("/".join(lst_string))
````

###  리스트 컴프리헨션(list comprehension) - 나중에 for문 할때 다시!  
*   리스트 재구성

*   각 요소에 같은 변화를 줘 리스트 전체를 바꾸는 방식

*   문법형식: [요소변화식 for 요소 in 리스트]



````python
# 리스트 컴프리헨션 (List Comprehension)
# → for문 없이 리스트를 재구성하는 간단한 방법

lst_numeric = [23, 52, 98, 31]

# (1) 각 요소에 100을 더해 새로운 리스트 생성
new_lst = [i + 100 for i in lst_numeric]
print(new_lst)   # [123, 152, 198, 131]

# (2) 각 요소를 2로 나눈 나머지를 요소로 하는 리스트 생성
lst_numeric_1 = [i % 2 for i in lst_numeric]
print(lst_numeric_1)   # [1, 0, 0, 1]


# (3) 출생연도 리스트 → 나이 리스트로 변환 (2024년 기준)
lst_year = [2004, 2006, 2004, 2008, 2005]
lst_age = [2024 - (i - 1) for i in lst_year]   # 만 나이 계산 예시
print(lst_age)   # [21, 19, 21, 17, 20]


# (4) 이름 뒤에 "안녕" 붙이기
lst_string = ["하니", "해린", "민지", "혜인", "다니엘"]
print([member + "안녕" for member in lst_string])
# ['하니안녕', '해린안녕', '민지안녕', '혜인안녕', '다니엘안녕']

# 조건문 추가: "다니엘"은 제외
print([member + "안녕" for member in lst_string if member != "다니엘"])

# 조건문 + else 추가
print([member + "안녕" if member != "다니엘" else "안녕하세요" for member in lst_string])


# (5) 이름의 길이가 3이 아닌 멤버만 추출
lst_string_1 = [i for i in lst_string if len(i) != 3]
print(lst_string_1)   # ['해린', '다니엘']
````

## 튜플(tuple)

* **소괄호 ( )** 로 감싸고, 각 요소는 쉼표(,)로 구분  
* 리스트와 비슷하지만, **수정·삭제 불가능 (immutable)**  
* 처리 속도가 리스트보다 빠르고, **데이터 변경이 필요 없는 경우 유용**  
* 예: 좌표, 요일, 고정된 데이터 집합 등





````python
tp_null = () # 비어있는 튜플 생성

tp_1 = (1,) #  요소가 하나만 튜플을 만들 때는 요소 뒤에 쉼표 (필요없음)

t1 = (1, 2, 3)

t2 = (1, "뉴진스", True)

tp_string = ("하니", "해린", "민지", "혜인", "다니엘")

tp_numeric = (162, 165, 169, 170, 165) # 다섯 개의 요소를 가진 튜플 (괄호를 생략해도 생성)

tp_numeric = 162, 165, 169, 170, 165
````

### 튜플 연산

* (문자열과 리스트처럼) 이어붙이기와 반복해서 입력 가능

````python
# 튜플 예시
tp_string = ("하니", "해린", "민지", "혜인", "다니엘")
tp_numeric = (162, 165, 169, 170, 165)

### 1. 튜플 연산
# (문자열/리스트와 마찬가지로 가능)

# 더하기 (concatenation)
print(tp_string + tp_numeric)
# ('하니', '해린', '민지', '혜인', '다니엘', 162, 165, 169, 170, 165)

# 곱하기 (repetition) → 원본은 변하지 않음
print(tp_string * 2)
# ('하니', '해린', '민지', '혜인', '다니엘','하니', '해린', '민지', '혜인', '다니엘')
````

### 튜플 인덱싱과 슬라이싱

````python
### 2. 튜플 인덱싱과 슬라이싱
print(tp_string[2])      # 민지
print(tp_string[1:5])    # ('해린', '민지', '혜인', '다니엘')
````

  ### 튜플 요소 수정하기(그 자체로는 불가능!!!)

````python
### 3. 튜플 요소 수정 (불가능!!!)
# 리스트와 달리 튜플은 immutable → 요소 수정/삭제 시 에러 발생
# 아래 코드는 모두 TypeError 발생 예시

del tp_string[4]         # TypeError
tp_numeric[0] = 163      # TypeError
tp_string[2] = "장원영"  # TypeError
````

````python
# 튜플 자체를 수정하려면? (우회 방법: 집에서 해볼것)
# 1) 튜플 → 리스트 변환
# 2) 리스트에서 요소 수정
# 3) 리스트 → 튜플 변환

tp_string = ("하니", "해린", "민지", "혜인", "다니엘")
print("원본 튜플:", tp_string)

# (1) 튜플을 리스트로 변환
lst_string = list(tp_string)

# (2) 리스트 요소 수정
lst_string[2] = "장원영"

# (3) 다시 튜플로 변환
tp_string = tuple(lst_string)

print("수정된 튜플:", tp_string)
# 원본 튜플: ('하니', '해린', '민지', '혜인', '다니엘')
# 수정된 튜플: ('하니', '해린', '장원영', '혜인', '다니엘')


print(type(lst_string))  # <class 'list'>
print(type(tp_string))   # <class 'tuple'>
````

## 사전(dictionary) - 딕셔너리

* **키(key)와 값(value)** 을 한 쌍으로 갖는 자료형 (예: `"키": 값`)

## 집합(set)

* 집합 자료형 생성: `set()`을 사용하여 생성 (문자열, 리스트, 튜플 등 시퀀스를 변환 가능)  
* 중괄호 `{ }` 로 감싸고, 각 요소는 쉼표(,)로 구분  
* 리스트와 유사하지만:  
  - **중복을 허용하지 않음**  
  - **순서가 없음(unordered)** → 인덱싱 불가  
* 원소의 **중복 제거**, **교집합/합집합/차집합 연산** 등에 유용

* 집합은 잘 안 쓰이는 것 같지만, 실제로는 중복 제거와 교집합·합집합 같은 연산 때문에 데이터 처리에서 많이 씁니다.



### 집합(set) 생성

````python
### 집합(set) 생성

# {} → 빈 집합이 아니라 빈 딕셔너리!
st_null = {}
print(type(st_null))   # <class 'dict'>

# 빈 집합 생성 방법
st = set()
print(type(st))        # <class 'set'>

# 숫자 리스트 → 집합 변환 (중복 자동 제거)
st1 = set([162, 165, 169, 170, 165])
print(st1)             # {162, 165, 169, 170}

# 문자열 리스트 → 집합 변환
st2 = set(["하니", "해린", "민지", "혜인", "다니엘"])
print(st2)

# 문자열 자체 → 문자열도 집합으로 변환 가능(문자 하나하나가 집합의 원소가 됨 > 중복은 삭제)
st3 = set("Minji")  # 참고: 구분자 없이 문자열을 list("Minji")로 변환하면 문자 하나하나가 리스트 객체의 원소가 됨!
print(st3)          # {'M', 'n', 'j', 'i'}


# 중괄호 { } 로 직접 생성 가능
st_numeric = {162, 165, 169, 170, 165}   # 객체 생성하면 바로 중복 삭제가능 > {162, 165, 169, 170} > mutable
st_string  = {"하니", "해린", "민지", "혜인", "다니엘"}
````

###   집합을 이용해 중복된 요소 정리



````python
lst_string = ["하니", "해린", "민지", "혜인", "다니엘", "하니"]

# 리스트 → 집합 변환 (중복 제거)
st_string_1 = set(lst_string)
print(st_string_1)     # {'다니엘', '민지', '해린', '혜인', '하니'}

# 집합 → 다시 리스트 변환
lst_string_1 = list(st_string_1)
print(lst_string_1)    # ['다니엘', '민지', '해린', '혜인', '하니']
````

### 집합 연산 (교집합, 합집합, 차집합)

* 덧셈, 뺄샘, 곱셉, 나눗셈 불가능
* 그래서 다른 방식(교집합, 합집합, 차집합, 집합의 메서드 등)으로 자료를 정리

````python
뉴진스 = {"하니", "해린", "민지", "혜인", "다니엘", "방시혁"}
BTS    = {"RM", "진", "슈가", "제이홉", "지민", "뷔", "정국", "방시혁"}

# 교집합 (공통 요소)
print(뉴진스 & BTS)    # {'방시혁'}

# 합집합 (전체 멤버)
print(뉴진스 | BTS)

# 차집합 (뉴진스에만 있는 멤버)
print(뉴진스 - BTS)

````

### 집합의 주요 메서드

* **요소 추가하기**
  - `add(x)` : 집합에 한 개의 요소 추가
  - `update([a, b, ...])` : 여러 요소 한꺼번에 추가 (리스트, 튜플, 집합 등 시퀀스 가능)

* **요소 제거하기**
  - `remove(x)` : 요소 삭제 (없으면 **KeyError** 발생)
  - `discard(x)` : 요소 삭제 (없어도 에러 발생하지 않음)
  - `clear()` : 집합의 모든 요소 제거


````python
# 초기 집합
st_string_2 = {"하니", "해린", "민지", "혜인", "다니엘"}

# (1) 요소 추가
st_string_2.add("카리나")             # 단일 요소 추가
print(st_string_2)

st_numeric_2 = {162, 165, 169, 170}
st_numeric_2.update([172, 175])      # 여러 요소 추가
print(st_numeric_2)

# (2) 요소 제거
st_string_2.remove("카리나")         # 요소 삭제
print(st_string_2)

# st_string_2.remove("없는값")       # KeyError 발생!
st_string_2.discard("없는값")         # 에러 없이 무시됨
print(st_string_2)

# (3) 전체 삭제
st_string_2.clear()
print(st_string_2)   # 빈 집합 set()
````

*  비교연산자(<, >, >=, <=, ==, !=) 이용

````python
# 변수 p1과 p2에 각각 "하니"와 "혜인"이라는 이름을 할당하고,
# 하니_생년과 혜인_생년에 각각 그들의 출생 연도를 할당한 후,
# 비교연산자(<, >, >=, <=, ==, !=)를 이용해 이 두 사람의 나이를 비교하여 서로의 관계를 출력하는 조건문을 작성


# 변수 선언
p1, p2 = "하니", "혜인"

# 나이 정보 선언
하니_출생연도, 혜인_출생연도 = 2004, 2008

# 조건문
if 하니_출생연도 == 혜인_출생연도:
    print(f"{p1}와 {p2}는 동갑입니다")
elif 하니_출생연도 < 혜인_출생연도:
    print(f"{p1}가 {p2}의 언니입니다")
else:
    print(f"{p1}가 {p2}의 동생입니다")
# f스트링 써야한다 헷갈리지 말자
````

*  논리연산자 이용

````python
# p1과 p2에 각각 "민지"와 "카리나"라는 이름을 할당한 후,
# 논리연산자를 이용해서 이 두 이름이 members 리스트에 포함되어 있는지를 확인하여,
# 그 결과에 따라 다른 메시지를 출력


members = ["하니", "해린", "민지", "혜인", "다니엘"]

p1, p2 = "민지", "카리나"

if (p1 in members) and (p2 in members):
    print(f"{p1}와 {p2}는 모두 뉴진스 멤버입니다")
elif (p1 in members) or (p2 in members):
    print(f"{p1}와 {p2}는 적어도 한명이 뉴진스 멤버입니다")
else:
    print(f"{p1}와 {p2}는 모두 뉴진스 멤버가 아닙니다")

````

## 제어문(2) - while문(반복문)

* 특정 조건이 **참(True)** 인 동안 코드 블록을 반복 실행  
* 조건이 **거짓(False)** 이 되면 루프 종료  
* 조건이 항상 참이라면 무한 루프 발생 → `break` 문으로 빠져나올 수 있음
*   while문 조건에 맞지 않을때 다시 처음부터 수행하고 싶을때는 continue를 사용




image.png

* 예시 1: 1부터 9까지 출력

````python
x = 1

while x < 10:
    print(x)
    x += 1   # 조건을 변화시켜야 루프 종료됨  x = x +1

while x < 10:  #위 아래가 다르다
    x += 1
    print(x)
````

* 예시 2: 무한 루프 + break

````python
while True:
    print("한 번만 출력")
    break   # 반복 종료
````

* 예시 3: continue 사용
    - 특정 조건일 때 **현재 반복을 건너뛰고**, 다음 반복을 진행
    - pass 와 continue의 다른 점 pass는 그냥 지나가서 밑에 실행문을 해라~ 이거고 continue는 이번 반복만 건너뛰어! 이거임
    - continue 하고 ':' 안쓴다!!!

````python
x = 0

while x < 5: # x = 0
    x += 1    # x = 1
    if x == 3:
        continue    # x가 3일 때는 print 실행 건너뜀
    print(x)

x = 0

while x < 5: # x = 0
    x += 1    # x = 1
    if x == 3:
        pass    # 결국 똑같이 3이 나온다. 즉, 1,2,3,4,5 이렇게 줄 출력된다. 물론 줄 바꿔서
    print(x)
````

* 딕셔너리와 for문
    - 딕셔너리는 **키(key)** 와 **값(value)** 으로 구성됨.  
    - for문과 함께 사용할 때는 보통 두 가지 방식이 자주 쓰임:

1. `.keys()` : 키만 꺼내 순회 → 각 키를 이용해 값을 조회  
2. `.items()` : 키-값 쌍을 동시에 꺼내 순회

````python
# 사전 자료형을  for 반복문을 통해
# 사전의 키를 순회하면서 키(각 국가)에 대해 연결된 값(수도)을 조회하고, 이를 출력하는 예제

# 예제 9: keys() 사용 → 키만 꺼내고, 값은 조회
country = {"한국": "서울", "중국": "베이징", "미국": "워싱턴"}

for ct in country.keys():
    print(f"{ct}의 수도는 {country[ct]}입니다")
# 출력:
# 한국의 수도는 서울입니다
# 중국의 수도는 베이징입니다
# 미국의 수도는 워싱턴입니다
````

````python
# 사전 자료형을 사용하여 for 루프와 items() 메서드를 사용하여
# 키-값 쌍을 반복문으로 순회하면서 각 국가와 그 수도를 출력하는 예제

# 예제 10: items() 사용 → 키-값 쌍을 동시에 unpack
for c, p in country.items():
    print(f"{c}의 수도는 {p}입니다")
# 출력:
# 한국의 수도는 서울입니다
# 중국의 수도는 베이징입니다
# 미국의 수도는 워싱턴입니다
````

### 예외 처리(try/except)
- **예외(Exception)**: 코드 실행 중에 발생하는 오류
- **프로그램이 중간에 멈추지 않도록, 에러를 처리하는 방법이 필요**
- `try` 블록에 에러가 발생할 수 있는 코드를 작성
- `except` 블록에 에러가 발생했을 때 실행할 코드를 작성

````python
#"뉴진스 멤버 확인 프로그램 (예외 처리 포함)"
# 리스트에서 특정 이름이 있는지 확인하면서,예외가 발생해도 프로그램이 멈추지 않고 계속 실행되도록 만든 예제

members = ["하니", "해린", "민지", "혜인", "다니엘"]

try: # 블록 안에서 코드를 실행 - 실행중 오류가 발생하면 멈추지 않고 excpet 블로으로 이동
    name = int(input("이름을 입력하세요: "))
    if name in members:
        print(f"{name}는 뉴진스 멤버입니다.")
    else:
        print(f"{name}는 뉴진스 멤버가 아닙니다.")
except Exception as e: # 모든 에러를 잡아 변수 e에 담는다는 뜻
    print("에러가 발생했습니다:", e)

try: # 블록 안에서 코드를 실행 - 실행중 오류가 발생하면 멈추지 않고 excpet 블로으로 이동 # 이 경우 에러가 안난다.
    name = input("이름을 입력하세요: "))
    if name in members:
        print(f"{name}는 뉴진스 멤버입니다.")
    else:
        print(f"{name}는 뉴진스 멤버가 아닙니다.")
except Exception as e: # 모든 에러를 잡아 변수 e에 담는다는 뜻
    print("에러가 발생했습니다:", e)
````

### 함수 정의 관련 용어
*   def: 함수 정의할때 사용하는 키워드

*   매개변수(parameter): 함수에 입력으로 전달될 값을 받는 변수

*   인자(arguments): 함수를 호출할 때 전달하는 입력값

*   수학에서 y = 2x + 1에서 x는 입력값(input), y는 출력값(output) > 혹은 x와 y는 독립변수, 종속변수라고 함

*   프로그래밍에서 y = 2*x + 1에서 x는 매개변수(parameter), y는  반환값(return value) > x와 y는 모두 함수안에서만 살아있는 지역변수(local variable)

*   프로그래밍에서 이 특정 함수명(예, my_func(x))를 호출하면서 x에 특정값(예, 3)을 할당하면 3을 인자라고 함.

````python
# 예제1: 함수의 기본 구조 1

def add(x, y): # def는 함수 정의할때 사용하는 키워드, add는 함수명,  x, y는 매개변수(입력변수), 초기값 설정가능
    result = x + y # 실행문:x + y -> result는 결과값
    return result  # 결과값을 반환하라
````

````python
# 예제1의 함수 호출해서 결과값 반환
# add() 함수 호출하고 괄호안 인자에 x=4000, y=5를 집어 넣음
add(4000, 5)

# add() 함수 다시 사용하고 이번에는 인자 3, 5를 집어 넣음,
add(x = 3, y = 5)
````

*   파이썬은 *args와 *kwargs 는 기능을 제공

*   여러 개의 입력값을 받는 함수를 사용할때는 매개변수를 *args로 표시



````python
# 여러 개의 입력값을 받는 매개변수 - *args > 가변인자 > 인자가 여러개

def multiple_add(*args):   #  *args:  매개변수 앞에 *을 붙이면 입력값을 전부 모아 튜플로 만들어 줌
    result= sum(args)
    return result

multiple_add(2,3,4)


````

````python
# 가변 인자(*args)를 사용하여 여러 이름을 받아들인 후,
# 각 이름이 특정 그룹(뉴진스)의 멤버인지 확인하는 함수 예제

members = ["하니", "해린", "민지", "혜인", "다니엘"] #여러 함수나 코드에서 사용하는 리스트

def 뉴진스_member_detect(*args):
    result = []
    for p in args:
        if p in members:
            result.append(f"{p}는 뉴진스 멤버입니다.")
        else:
            result.append(f"{p}는 뉴진스 멤버가 아닙니다.")
    return result

뉴진스_member_detect("하니", "카리나", "다니엘")
# 여기서 리턴값이 ['하니는 뉴진스 멤버입니다.', '카리나는 뉴진스 멤버가 아닙니다.', '다니엘는 뉴진스 멤버입니다.']와 같은 리스트로 나온다는 것을 주의하자
# 이는 당연하다 result가 리스트이기 때문에 return result 하면 당연히 리스트로 나온다
````



*   딕셔너리 형식의 입력값을 받는 함수를 정의하고 호출하려면 **kwargs(키워드 파라미터) 사용





````python
# 가변 키워드 인자(**kwargs)를 사용하여 여러 나라와 그 나라의 수도를 입력받아,
# 각 나라의 수도를 설명하는 문장을 생성하고
# 그 결과를 리스트로 반환하는 것

# **kwargs 보면 결국 입력부분을 사전처럼 만들겠다는 것을 알아야함

def country_capital(**kwargs): #**kwargs를 사용해서 함수 호출 시 여러 나라와 그 나라의 수도를 키-값 쌍의 형태로 입력받을 것임
    result = []
    for key, value in kwargs.items():
        text = f"{key}의 수도는 {value}입니다" # 문장 생성
        result.append(text) # 생성된 모든 문장을 리스트에 담아라
    return result # 리스트를 반환하라
country_capital(한국="서울", 중국="베이징") #호출하면, kwargs는 {'한국': '서울', '중국': '베이징'}라는 딕셔너리가 됨


````

✅ 1. 함수 호출에서 = 을 쓰는 이유

country_capital(한국="서울", 중국="베이징")

이 부분을 보면 딕셔너리처럼 보이지만, 실제로는 딕셔너리를 직접 쓰는 문법이 아니기 때문이에요.

📌 왜 = 을 쓰는가?

함수 호출 시 사용하는 문법은 키워드 인자(keyword argument) 문법이에요.

키워드 인자를 전달할 때는

함수이름(키=값)


형태를 사용해야 해요.

즉,
함수 호출의 문법은 딕셔너리 문법과 다르다!

그래서 함수 호출에서는 : 를 쓰지 않아요.

🔍 그럼 내부적으로는 어떻게 될까?

파이썬은 내부에서 이 키워드 인자들을 자동으로 딕셔너리로 바꿔서 kwargs로 전달해요.

즉, 아래 두 단계가 일어나는 것:

우리가 입력 → 한국="서울"

파이썬 내부에서 변환 → {"한국": "서울"}

그래서 함수 내부에서는 딕셔너리로 처리하게 되는 거예요.

✅ 2. 왜 key에는 따옴표를 안 쓰는가?

함수 호출 문법에서
한국="서울"
여기서 **한국은 문자열(string)이 아니고 '변수 이름처럼 취급되는 식별자(identifier)'**입니다.

📌 파이썬 키워드 인자의 규칙

key는 반드시 식별자(identifier)이어야 함

즉, 알파벳·숫자·언더바로 이루어진 변수명처럼 생겨야 함

"한국"이라는 문자열은 식별자가 아니기 때문에 사용할 수 없음

그래서 아래는 오류가 나는 코드:

country_capital("한국"="서울")  # ❌ 문법 오류


반면 "서울"은 함수에 전달되는 값(value) 이므로 문자열로 적어야 해요.

# 클래스(class)

* 객체지향 프로그래밍(OOP)의 핵심 개념
* 클래스(Class): 같은 구조와 기능을 가진 객체를 반복해서 만들 수 있는 **설계도(붕어빵 틀)**
* 객체(object) / 인스턴스(instance): 클래스로부터 생성된 **실제 대상(붕어빵)**
* 메서드(method): 클래스 안에 정의된 함수 → 객체가 수행할 수 있는 동작
* 생성자(constructor): 객체 생성 시 자동으로 호출되는 메서드 (`__init__`)  
  → 객체의 초기값 설정에 사용


* 클래스 기본구조

````python
class 클래스이름:
    def __init__(self, 매개변수):
        # 생성자 (constructor)
        self.속성 = 매개변수

    def 메서드이름(self):
        # 메서드 (method)
        동작 코드
````

## 클래스의 구조 이해하기

* 객체(인스턴스) 생성하기 = 설계도로 만든 제품 만들기
* 위에서 정의한 클래스를 사용해서 특정한 객체 생성
    - 예) 붕어빵 틀에 반죽을 부어 붕어빵 하나를 만드는 것
* 클래스를 이용해 객체(인스턴스)를 만들 때는, 보통 변수에 저장해서 사용

````python
# 예제1: 간단한 계산기 클래스를 정의한 예제 > 계산기 설계도
# 이 코드는 "Calculation"이라는 클래스를 정의하고 있음
# 이 클래스는 두 숫자를 입력받아 더하기, 빼기, 곱하기, 나누기 같은 기본적인 계산을 할 수 있는 기능을 가지고 있음

class Calculation:                      # 클래스는 대문자로("C"aculation), 객체는 소문자로
    def __init__(self, n1, n2):         #  __init__(self, n1, n2)는 생성자(constructor) 표시.
        self.n1 = n1                    # 아래에 객체가 생성될 때 이 생성자가 자동으로 호출됨. 객체 속성의 초기화작업 담당
        self.n2 = n2                    # 객체가 생성될 때, 전달된 두 개의 값 n1과 n2를 해당 객체의 속성 self.n1과 self.n2에 저장하여 초기화합니다.
                                        # self는 아래 생성될 객체 자신을 가리킴, 이를 통해 클래스 속성에 접근할 수 있음

    def add(self):                      # 객체가 수행할 동작인 2개의 함수(메서드) 정의: 더하기, 곱하기
        result = self.n1 + self.n2
        return result

    def mul(self):
        result = self.n1 * self.n2
        return result

````

````python
# 예제2: 클래스를 이용해서 특정 속성(8. 5)을 가진 객체(cal1) 생성하기
cal1 = Calculation(8, 5) # 이때 클래스의 생성자인 __init__ 메서드를 자동으로 호출

# # cal1 객체인 속성인 cal2.n1와 cal2.n2는?
cal1.n1 # 객체의 속성 8
cal1.n2 # 객체의 속성 5

# <클래스, 객체, 속성, 메서드에 대한 설명>
# Calculation 클래스라는 설계도를 사용해 'cal1'이라는 계산기 객체를 하나 만듬
# 'cal1' 객체는 8과 5를 가지고 태어나며(=객체의 속성)
# 클래스의 _init__ 메서드를 자동으로 호출하면서 n1과 n2에 각각 8과 5가 전달되며 이 값들은 self.n1과 self.n2에 저장
# 즉, n1과 n2를 받아서, 객체(cal)의 속성인 self.n1과 self.n2를 초기화하는 역할
# 클래스의 메서드인 add()과 mul()같은 동작을 수행할 수 있음
````

````python
# 파이썬으로 웹 크롤링을 수행하여 Yes24 베스트셀러 페이지에서 책 제목을 가져와서 화면에서 출력만 하는 예시

import requests
from bs4 import BeautifulSoup

# URL 변수에 크롤링할 Yes24 베스트셀러 페이지의 URL을 저장
URL ='https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber=1&pageSize=24'

# 해당 웹 페이지의 HTML 데이터를 요청하여 가져
html = requests.get(URL)

 #*html.text**로 받은 HTML 코드 전체를 파싱(분석하여 필요한 부분을 추출)
soup = BeautifulSoup(html.text,'html.parser') # 'html.parser': HTML 코드를 이해할 수 있는 포맷으로 변환하는 도구


books = soup.find_all('div', class_='itemUnit') # HTML에서 div 태그 중에서 클래스가 'itemUnit'인 부분을 모두 찾아서 books 리스트에 저장. div.itemUnit 부분에는 각 책의 정보가 포함
for book in books:
    print(book.find('a', class_='gd_name').text) # 각 책에서 책 제목이 들어있는 a 태그(클래스 'gd_name')를 찾아 그 텍스트(책 제목)를 추출
````

* 엑셀의 표(스프레드시트)처럼 행(Row)과 열(Column)로 구성된 비슷한 자료구조

* 행은 각각의 **관측값(레코드)**, 열은 각각의 **특성(변수)**를 의미

* 파이썬 작업환경에서 엑셀과 같은 표 데이터를 쉽게 저장·분석·가공할 수 있도록 해줌

* 특히 크롤링 결과(주로 리스트, 딕셔너리 등)를 DataFrame으로 변환하면, 곧바로 CSV, 엑셀파일로 손쉽게 저장가능

* 즉, 데이터프레임은 파이썬 자료형과 CSV나 엑셀과 같은 스프레드 시트 파일 사이를 연결해주는 다리이자, 파이썬에서 표 데이터를 다루는 핵심 도구임

````python
# Yes24 베스트셀러 페이지에서 책 제목, 저자, 출간년도, 출판사를 추출해 Pandas 데이터프레임으로 정리하는 코드

import pandas as pd

# 데이터 저장을 위한 리스트 초기화
data = []

# 책 정보 리스트에서 데이터 추출 후 리스트에 저장
for book in books:
    # 서적 이름 추출
    title_tag = book.find('a', class_='gd_name')
    title = title_tag.get_text(strip=True) if title_tag else "N/A"

    # 저자명 추출
    author_tag = book.find('span', class_='authPub info_auth')
    author = author_tag.find('a').get_text(strip=True) if author_tag and author_tag.find('a') else "N/A"

    # 출간년도 추출
    publication_date_tag = book.find('span', class_='authPub info_date')
    publication_date = publication_date_tag.get_text(strip=True) if publication_date_tag else "N/A"

    # 출판사명 추출
    publisher_tag = book.find('span', class_='authPub info_pub')
    publisher = publisher_tag.find('a').get_text(strip=True) if publisher_tag and publisher_tag.find('a') else "N/A"

    # 데이터를 리스트에 추가
    data.append([title, author, publication_date, publisher])

# Pandas 데이터프레임 생성
df = pd.DataFrame(data, columns=['서적 이름', '저자', '출간년도', '출판사'])

# 데이터프레임 앞 부분만 미리 확인가능
df.head()

# 전체데이터프레임 출력
print(df)
````
